```python
def PUSH(stack, data, max):
    if (len(stack) >= max):
      print("Stack FULL")
      return
    stack.append(data)
    return
 
def POP(stack):
    if (len(stack) == 0):
      print("Stack Empty")
      return
    stack.pop()
    if (len(stack) == 0):
      print("Stack Empty")
      return
    return

def PEEK(stack):
    if (len(stack) == 0):
      print("Stack Empty")
      return
    data = stack[-1]
    print(data)
    if (len(stack) == 0):
      print("Stack Empty")
      return
    return

def DUP(stack, max):
    if (len(stack) == 0):
      print("Stack Empty")
      return
    data = stack.pop()
    if (len(stack) >= max):
      print("Stack FULL")
      return
    stack.append(data)
    if (len(stack) >= max):
      print("Stack FULL")
      return
    stack.append(data)

def UpR(stack, n, max):
    if (len(stack) == 0):
      print("Stack Empty")
      return
    if (n > max):
      return
    tmp = stack[len(stack)-1]
    i = 0
    for i in range(1, n):
      stack[len(stack) - i] = stack[len(stack) - (i+1)]
    stack[len(stack) - (n)] = tmp
    return

def DownR(stack, n, max):
    if (len(stack) == 0):
      print("Stack Empty")
      return
    if (n > max):
      return
    tmp = stack[len(stack) - n]
    i = 0
    for i in range(1, n):
      stack[len(stack) - (n - i + 1)] = stack[len(stack) - (n - i)]
    stack[len(stack)-1] = tmp
    return

def PRINT(stack):
    i = 0
    for i in range(0, len(stack)):
      print(stack[len(stack)-(i+1)], end='')
    print('')
    return
  

 
stack = list()
max = int(input())
n = int(input())
for i in range(0, n):
  k = list(input().split())

  if (k[0] == 'PUSH'):
    PUSH(stack, k[1], max)
  elif (k[0] == 'POP'):
    POP(stack)
  elif (k[0] == 'DUP'):
    DUP(stack, max)
  elif (k[0] == 'PEEK'):
    PEEK(stack)
  elif (k[0] == 'UpR'):
    UpR(stack, int(k[1]), max)
  elif (k[0] == 'DownR'):
    DownR(stack, int(k[1]), max)
  elif (k[0] == 'PRINT'):
    PRINT(stack)
  else:
    break
```
