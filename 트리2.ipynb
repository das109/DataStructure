```python
class Node:
    def __init__(self, data, id):
      self.data = data
      self.id = id
      self.left = None
      self.right = None
    def __str__(self):
        return str(self.data)

class Tree:
  def __init__(self):
      self.root = None

  def addNode(self, node, left_node, right_node):
      if self.root == None:
          self.root = node
      node.left = left_node
      node.right = right_node


  def Search(self, Node, order, id):
      now_node = Node
      if (now_node.left != None) and (now_node.left.data is not None): 
          self.Search(now_node.left, order, id)
      if (now_node.right != None) and (now_node.left.data is not None): 
          self.Search(now_node.right, order, id)
      if (now_node.id == id):
          node = now_node
          if (order == 1):
            #print('', end=' ')
            #print(node.data, end=' ')
            self.Preorder(node)
          elif (order == 2):
            #print('', end=' ')
            #print(node.data, end=' ')
            self.Inorder(node)
          elif (order == 3):
            #print('', end=' ')
            #print(node.data, end=' ')
            self.Postorder(node)

  def Preorder(self, node):
      print(node.data, end=' ')
      if (node.left != None) and (node.left.data is not None):
        self.Preorder(node.left)
      if (node.right != None) and (node.right.data is not None):
        self.Preorder(node.right)   

  def Inorder(self, node):
      if (node.left != None) and (node.left.data is not None): 
        self.Inorder(node.left)
      print(node.data, end=' ')
      if (node.right != None) and (node.right.data is not None): 
        self.Inorder(node.right)
    
  def Postorder(self, node):
      if (node.left != None) and (node.left.data is not None): 
        self.Postorder(node.left)
      if (node.right != None) and (node.right.data is not None): 
        self.Postorder(node.right)
      print(node.data, end=' ')
    

li_tree = ['*', Node(20, 1), Node(30, 2), Node(50, 3), Node(70, 4), Node(90, 5), Node(120, 6), Node(130, 7), Node(80, 8)]

tree = Tree()
tree.addNode(li_tree[1], li_tree[2], li_tree[3])
tree.addNode(li_tree[2], li_tree[4], li_tree[5])
tree.addNode(li_tree[3], li_tree[6], None)
tree.addNode(li_tree[6], li_tree[7], li_tree[8])



order = input()
id = input()
if (int(id) >= len(li_tree)):
  print(-1)

else:
  tree.Search(tree.root, int(order), int(id))
```
