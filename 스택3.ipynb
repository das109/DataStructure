```python
def preorder(word):
  #if (word == '!') or (word == '+') or (word == '-'):
  if (word == '!'):
    check = 6
  elif (word == '*') or (word == '/'):
    check = 5
  elif (word == '+') or (word == '-'):
    check = 4
  elif (word == '>') or (word == '<'):
    check = 3
  elif (word == '&'):
    check = 2 
  elif (word == '|'):
    check = 1
  else:
    check = 0
  
  return check


n = int(input())

for i in range(0, n):
  string = str(input()) #문자열을 리스트로 저장
  Slist = list(string)
  result = []
  stack = []
  a = len(Slist)
  flag = 0

  for i in Slist: 
    if ((i == '&') or (i == '|')) and (flag == 0):
      flag = 1
      continue
    if ((i == '&') or (i == '|')) and (flag == 1):
      flag = 0
      pass
    
    check = preorder(i)

    if (i == '('):
      stack.append(i)
    elif (i == ')'):
      while (1):
        a = stack.pop()
        #print('a는 '+str(a))
        if (a == '('):
          break
        result.append(a)

    elif (check == 0):
      result.append(i)

    else:
      #print("스택은")
      #print(stack)
      for j in range(0,len(stack)):
        op = stack.pop()
        #print("check는 "+str(check))
        #print("op는 "+str(preorder(op)))
        if (op == '(') or (op == ')'):
          stack.append(op)
          break
        if (preorder(op) < check):
          stack.append(op)
          break
        result.append(op)
        #print("op는 "+str(op))
      stack.append(i)
  for j in range(0,len(stack)):
    op = stack.pop()
    result.append(op) 
    
  for i in result:
    if (i == '&'):
      print("&&", end='') 
    elif (i == '|'):
      print("||", end='') 
    else:
      print(i, end ='')
  print("")
```
